```python
import pandas as pd
```

```python
folder_output = '/mbshome/nvelthuijs/Cofactors/20200402_Transcriptionregulationlist/Output_files/'

list_file = folder_output + '16_Transcription_regulatorlist_baitset3_with_scores_with_blacklist.csv'

df = pd.read_csv(list_file)
```

```python
mf_terms = ['GO:0003899',
           'GO:0003711',
           'GO:0140223',
           'GO:0000981',
           'GO:0003700',
           'GO:0003712',
           'GO:0003713',
           'GO:0003714']

bp_terms = ['GO:0006366',
           'GO:0006352',
           'GO:0006367',
           'GO:0006355',
           'GO:0006357']

cc_terms = ['GO:0005665']

def get_termcounts(termslist, df_in, namespace):
    counts_dict = {}
    
    for term in termslist:
        counts_dict[term] = 0

    for row in df_in['{} terms'.format(namespace)]:
        if isinstance(row, str):
            row = set(row[1:-1].split(sep = ', '))
        else:
            continue
            
        for term in row:
                term = term[1:-1]
                if term in counts_dict:
                    counts_dict[term] += 1
                
    return counts_dict

mf_counts = get_termcounts(mf_terms, df, 'MF')
bp_counts = get_termcounts(bp_terms, df, 'BP')
cc_counts = get_termcounts(cc_terms, df, 'CC')
all_counts = {**mf_counts, **bp_counts, **cc_counts}

df_newfound = df.loc[((df['In_bait_crems'] == 0) &
                (df['In_bait_snfs'] == 0) &
                (df['In_bait_nursa'] == 0) &
                (df['In_bait_gocofs'] == 0) &
                (df['Is_TF'] == 0) &
                (df['In_NVS'] == 0)
                )]

mf_counts = get_termcounts(mf_terms, df_newfound, 'MF')
bp_counts = get_termcounts(bp_terms, df_newfound, 'BP')
cc_counts = get_termcounts(cc_terms, df_newfound, 'CC')
all_counts2 = {**mf_counts, **bp_counts, **cc_counts}

for key in all_counts.keys():
    all_counts[key] = [all_counts[key], all_counts2[key]]

df_counts = pd.DataFrame.from_dict(all_counts, orient = 'index', columns = ['All proteins', 'Newfound proteins'])
df_counts['Bait proteins'] = df_counts['All proteins'] - df_counts['Newfound proteins']

folder_output = '/mbshome/nvelthuijs/Cofactors/20200402_Transcriptionregulationlist/Output_files/'
df_counts.to_csv(folder_output + '17_Quintes_GO_termcounts.csv')
df_counts
```
