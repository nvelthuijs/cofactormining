```python
with open('/mbshome/nvelthuijs/Cofactors/20200402_Transcriptionregulationlist/Datafiles/tfs.txt') as tfs_file:
    tfs = set([tf.strip() for tf in tfs_file])

with open('/mbshome/nvelthuijs/Cofactors/20200402_Transcriptionregulationlist/Datafiles/gocofs.txt') as gocof_file:
    gocofs = set([cotf.strip() for cotf in gocof_file])

gocofs_notfs = [cotf for cotf in gocofs if cotf not in tfs]

with open('/mbshome/nvelthuijs/Cofactors/20200402_Transcriptionregulationlist/Datafiles/gocofs_notfs.txt', 'w') as outfile:
    for cotf in gocofs_notfs:
        outfile.write(cotf + '\n')
```

```python
with open('/mbshome/nvelthuijs/Cofactors/20200402_Transcriptionregulationlist/Datafiles/golden_standard_proteins.txt') as pcs_file:
    pcs1 = [line.strip().split('\t', 1) for line in pcs_file]
pcs = {key:value.split('\t') for key, value in pcs1}
        
with open('/mbshome/nvelthuijs/Cofactors/20200402_Transcriptionregulationlist/Datafiles/golden_standard_complexes.txt') as pcs_file:
    complexes = [line.strip().split('\t', 1) for line in pcs_file]
complexes = {key:value.split('\t') for key, value in complexes}

for protein in pcs.keys():
    if protein in tfs:
        cnames = pcs[protein]
        for cname in cnames:
            print(protein, '\n', cname, '\n', complexes[cname], '\n')

with open('/mbshome/nvelthuijs/Cofactors/20200402_Transcriptionregulationlist/Datafiles/golden_standard_proteins_notfs.txt', 'w') as outfile:
    for protein, compl in pcs1:
        if not protein in tfs:
            outfile.write(protein + '\t' + compl + '\n')
```
